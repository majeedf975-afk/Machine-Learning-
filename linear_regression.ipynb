```python
import kagglehub
path = kagglehub.dataset_download("andonians/random-linear-regression")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model

```

```python
df = pd.read_csv(path + "/train.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(df.y,df.x)
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[['x']], df[['y']])
```

```python
lr.predict([[14]])
```

```python
lr.intercept_
```

```python
lr.score(df[['x']], df.y)
lr.coef_
```

```python
plt.xlabel('x',color='red')
plt.ylabel('y',color = 'red')
plt.scatter(df.y,df.x)
plt.plot(df.x, lr.predict(df[['x']]), color='red')
```
